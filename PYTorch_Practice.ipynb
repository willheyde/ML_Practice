```python
#PyTorch tensors and Numpy
#often to have data in Numpy and want/need it in Pytorch tensor
#you can do `torch.from_numpy(ndarray)`
#pytorch to numpy you can do torch.tensor.numpy
import numpy as pd
array =np.arange(0.0,11.0)
tensor = torch.from_numpy(array).type(torch.float32)
print(f"{array}")
print(f"{tensor}")
#When going from nump to pytoch, pytorch reflects numpys default datatype of float64 unless specified
array = array + 1
print(f"{array}\n{tensor}")
#going to tensor back to numpy
tensor = torch.ones(5).type(torch.int32)
array = tensor.numpy()
tensor = tensor + 1
tensor,array
```

```python
#Reproducability: trying to take the random out of random
#To reduce randomness in neural networks and pytorch comes the comcept of random seed
#random seed  "flavors" The randomness.
rand1 = torch.rand(3,4)
rand2 = torch.rand(3,4)
print(rand1 == rand2)
```

```python
#sets random seed 
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
rand3 = torch.rand(3,4)
torch.manual_seed(RANDOM_SEED)
rand4 = torch.rand(3,4)
print(rand3 == rand4)
```
